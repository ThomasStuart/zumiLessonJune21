```python
# Max vowels class. given a list of stings, traverse through a string and count the number of vowels. 
# return the string with the most vowels.

# ex1.)
#  “hello” -> 2 vowels (e,o) 
#  “world” -> 1 vowels (o) 

# return “hello, 2”



def getNumberVowels( word ): 
    counter = 0
    #vowels  = #list containin the vowels
    
    #hint: use a for-loop , go char by char to check if particualr char in word is vowel
    for i in range(0, len(word) ):
        char = word[i]
        
        # if what ? char is a vowel -> 
        if char == 'a' or char == 'e' or char == 'i' or char == 'o' or char == 'u':
            counter += 1 
            
    return counter 



print("hello->", getNumberVowels("hello") )# -> 2
print("----")
print("world->", getNumberVowels("world") )# -> 1 
```

```python
from math import inf 

def getNumberVowels( word ): 
    counter = 0
    #vowels  = #list containin the vowels
    
    #hint: use a for-loop , go char by char to check if particualr char in word is vowel
    for i in range(0, len(word) ):
        char = word[i]
        
        # if what ? char is a vowel -> 
        if char == 'a' or char == 'e' or char == 'i' or char == 'o' or char == 'u':
            counter += 1 
            
    return counter 



            # words1 = "hello", world 
            # words2 = "it", "dictionary", "food"
def mapWordsToCount(words):
    # creates an list called pairs 
    pairs = []
    
    # calcs the number of vowels per word
    for i in range(0, len(words)): 
        countV = getNumberVowels(words[i])
        pairs.append( (words[i], countV) )
    
    
    # selects the word with highest vowel count
    maxWord        = ""
    maxVowelCount  = -inf
    for i in range(0, len(pairs) ):
        currWord       = pairs[i][0]
        currVowelCount = pairs[i][1]
        
        # check if new max
        if currVowelCount > maxVowelCount:
            maxWord       = currWord
            maxVowelCount = currVowelCount 
        
        
    return maxWord, maxVowelCount 

    
words1 = ["hello", "world"]
print( mapWordsToCount( words1 ) )  # hello, 2

words2 = ["it", "dictionary", "food"]  
print( mapWordsToCount( words2 ) ) # dictionary, 4 
```
